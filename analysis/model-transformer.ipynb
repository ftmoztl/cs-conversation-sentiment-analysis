## DistilBERT classifier

Use a pretrained DistilBERT model for 3-class sentiment classification and fine-tune it.

The model used in the training notebook is: DistilBertForSequenceClassification --> This is a DistilBERT model with a classification head added on top for sentiment prediction.


Why DistilBERT?

DistilBERT is an encoder-only transformer. For text classification, it reads the whole input sequence, builds contextual representations, and then uses a classification head on top to predict the label.

DistilBERT is a smaller, faster version of BERT. It is a pretrained transformer model that understands language context and can be fine-tuned for classification tasks such as sentiment analysis.

So, DistilBERT:
- sentiment is mainly expressed in the conversation text
- captures contextual meaning better than traditional models
- is lighter and faster than full BERT
- the dataset is relatively small, use pretrained model for it


Alternatives to DistilBERT:

- BERT-base-uncased: Stronger than DistilBERT, but heavier and slower.
- RoBERTa-base: Often performs very well on classification tasks, but is also heavier than DistilBERT.
- ALBERT: Smaller and more parameter-efficient, but sometimes less straightforward to fine-tune.
- DeBERTa-v3-base: Very strong for classification, but heavier and more demanding.
- MiniLM: Smaller and faster than DistilBERT, good if resources are limited.

So, they can be heavy for this simple task.

- I'll keep DistilBERT as the main model --> baseline model
- MAYBE BERT or RoBERTa can be alternatives. If needed, we can compare with the baseline but

```python
import torch
from torch.optim import AdamW
from torch.utils.data import Dataset, DataLoader
from transformers import DistilBertForSequenceClassification, get_linear_schedule_with_warmup # from Hugging Face
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, classification_report
```

```python
if torch.backends.mps.is_available():
    device = torch.device("mps")
elif torch.cuda.is_available():
    device = torch.device("cuda")
else:
    device = torch.device("cpu")

print("Using device:", device) ## Because my device is Apple M1 with 16 GB memory
```

I use the same DistilBERT checkpoint as in preprocessing so the tokenization and model vocabulary remain aligned.

```python
model_checkpoint = "distilbert-base-uncased"
num_labels = 3

model = DistilBertForSequenceClassification.from_pretrained(
    model_checkpoint,
    num_labels=num_labels
)
model.to(device)
```

Model:

### Datasets and dataloaders (For PyTorch)

```python
class SentimentDataset(Dataset):
    def __init__(self, input_ids, attention_mask, labels):
        self.input_ids = torch.tensor(input_ids, dtype=torch.long)
        self.attention_mask = torch.tensor(attention_mask, dtype=torch.long)
        self.labels = torch.tensor(labels, dtype=torch.long)

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        return {
            "input_ids": self.input_ids[idx],
            "attention_mask": self.attention_mask[idx],
            "labels": self.labels[idx]
        }
```

```python
train_dataset = SentimentDataset(X_train_input_ids, X_train_attention_mask, y_train)
val_dataset = SentimentDataset(X_val_input_ids, X_val_attention_mask, y_val)
test_dataset = SentimentDataset(X_test_input_ids, X_test_attention_mask, y_test)

batch_size = 8

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

print("Train batches:", len(train_loader))
print("Validation batches:", len(val_loader))
print("Test batches:", len(test_loader))
```

A small batch size is used because transformer models require substantial memory.

### Optimizer and scheduler

```python
epochs = 10
learning_rate = 2e-5

optimizer = AdamW(model.parameters(), lr=learning_rate)
total_steps = len(train_loader) * epochs

scheduler = get_linear_schedule_with_warmup(
    optimizer,
    num_warmup_steps=0,
    num_training_steps=total_steps
)
```

A low learning rate is chosen because we are fine-tuning a pretrained model rather than training a new model from scratch.
This is a common small learning rate for transformer fine-tuning.

Don't forget, scheduler makes LR like:  current_lr = initial_lr * (1 - step / total_steps)

Means that: learning rate is not fixed by epoch, it decreases every batch

IMPORTANT NOTE: Because DistilBERT is a pretrained transformer rather than a randomly initialized model, only a small number of epochs is typically required for fine-tuning. Given the relatively small dataset size, training for too many epochs would likely cause overfitting rather than improved generalization.


```python
wandb.init(
    project="cs-conversation-sentiment-analysis",
    name="distilbert-transformer-baseline",
    config={
        "model_checkpoint": model_checkpoint,
        "num_labels": num_labels,
        "batch_size": batch_size,
        "epochs": epochs,
        "learning_rate": learning_rate,
        "device": str(device),
        "train_size": len(train_dataset),
        "val_size": len(val_dataset),
        "test_size": len(test_dataset)
    }
)
```

### Handle class imbalance with weighted loss

```python
class_counts = np.bincount(y_train) #counts how many samples belong to each class in the training set
class_weights = len(y_train) / (len(class_counts) * class_counts)
class_weights = torch.tensor(class_weights, dtype=torch.float).to(device)

loss_fn = torch.nn.CrossEntropyLoss(weight=class_weights)

print("Class counts:", class_counts)
print("Class weights:", class_weights)
```

Weighted loss is used because the classes are imbalanced, especially the positive class.

### Training and evaluation functions

```python
def train_one_epoch(model, data_loader, optimizer, scheduler, device, loss_fn):
    model.train()
    total_loss = 0

    for batch in data_loader:
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        labels = batch["labels"].to(device)

        optimizer.zero_grad()
        outputs = model(input_ids=input_ids, attention_mask=attention_mask)
        logits = outputs.logits
        loss = loss_fn(logits, labels)

        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()
        scheduler.step()

        total_loss += loss.item()

    return total_loss / len(data_loader)
```

```python
def evaluate_model(model, data_loader, device, loss_fn):
    model.eval()
    total_loss = 0
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for batch in data_loader:
            input_ids = batch["input_ids"].to(device)
            attention_mask = batch["attention_mask"].to(device)
            labels = batch["labels"].to(device)

            outputs = model(input_ids=input_ids, attention_mask=attention_mask)
            logits = outputs.logits
            loss = loss_fn(logits, labels)

            total_loss += loss.item()
            preds = torch.argmax(logits, dim=1)

            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    avg_loss = total_loss / len(data_loader)
    accuracy = accuracy_score(all_labels, all_preds)
    precision, recall, f1, _ = precision_recall_fscore_support(
        all_labels,
        all_preds,
        average="macro",
        zero_division=0
    )

    return avg_loss, accuracy, precision, recall, f1, all_labels, all_preds
```

### Train the model

```python
history = []
best_val_f1 = 0
best_model_path = "best_distilbert_model.pt"

for epoch in range(epochs):
    train_loss = train_one_epoch(model, train_loader, optimizer, scheduler, device, loss_fn)
    val_loss, val_acc, val_precision, val_recall, val_f1, _, _ = evaluate_model(model, val_loader, device, loss_fn)

    history.append({
        "epoch": epoch + 1,
        "train_loss": train_loss,
        "val_loss": val_loss,
        "val_accuracy": val_acc,
        "val_precision_macro": val_precision,
        "val_recall_macro": val_recall,
        "val_f1_macro": val_f1
    })

    print(f"Epoch {epoch + 1}/{epochs}")
    print(f"  Train loss: {train_loss:.4f}")
    print(f"  Val loss: {val_loss:.4f}")
    print(f"  Val accuracy: {val_acc:.4f}")
    print(f"  Val macro F1: {val_f1:.4f}")

    wandb.log({
        "epoch": epoch + 1,
        "train_loss": train_loss,
        "val_loss": val_loss,
        "val_accuracy": val_acc,
        "val_precision_macro": val_precision,
        "val_recall_macro": val_recall,
        "val_f1_macro": val_f1
    })

    if val_f1 > best_val_f1:
        best_val_f1 = val_f1
        torch.save(model.state_dict(), best_model_path)
        wandb.run.summary["best_val_f1"] = best_val_f1
        print("  Best model updated and saved.")
```

### Evaluate the best model

* Accuracy shows the overall proportion of correct predictions. BUT accuracy alone is not enough for this dataset because the classes are imbalanced.
* Precision measures how many predicted samples of a class are actually correct.
* Recall measures how many real samples of a class the model successfully finds.
* F1 combines precision and recall into one score.
* Use macro metrics in this evaluation. Macro means the metric is calculated separately for each class. Then the class-level scores are averaged equally.This gives negative, neutral, and positive the same importance.

Use Macro Precision, Macro Recall, and Macro F1 because the positive class is very small. Macro metrics prevent the model from looking good only because it performs well on the larger classes.

I choose the best model by Macro F1 because it balances precision and recall.

It is more suitable than accuracy for imbalanced multi-class classification.

It helps ensure that all sentiment classes are considered during model selection.

```python
best_model = DistilBertForSequenceClassification.from_pretrained(model_checkpoint, num_labels=num_labels)
best_model.load_state_dict(torch.load(best_model_path, map_location=device))
best_model.to(device)
```

```python
val_loss, val_acc, val_precision, val_recall, val_f1, val_labels, val_preds = evaluate_model(best_model, val_loader, device, loss_fn)
test_loss, test_acc, test_precision, test_recall, test_f1, test_labels, test_preds = evaluate_model(best_model, test_loader, device, loss_fn)

print("Validation metrics")
print(f"Loss: {val_loss:.4f}")
print(f"Accuracy: {val_acc:.4f}")
print(f"Macro Precision: {val_precision:.4f}")
print(f"Macro Recall: {val_recall:.4f}")
print(f"Macro F1: {val_f1:.4f}")

print("\nTest metrics")
print(f"Loss: {test_loss:.4f}")
print(f"Accuracy: {test_acc:.4f}")
print(f"Macro Precision: {test_precision:.4f}")
print(f"Macro Recall: {test_recall:.4f}")
print(f"Macro F1: {test_f1:.4f}")

wandb.log({
    "final_val_loss": val_loss,
    "final_val_accuracy": val_acc,
    "final_val_precision_macro": val_precision,
    "final_val_recall_macro": val_recall,
    "final_val_f1_macro": val_f1,
    "test_loss": test_loss,
    "test_accuracy": test_acc,
    "test_precision_macro": test_precision,
    "test_recall_macro": test_recall,
    "test_f1_macro": test_f1
})
```

```python
label_names = ["negative", "neutral", "positive"]

print("Validation classification report:")
print(classification_report(val_labels, val_preds, target_names=label_names, zero_division=0))

print("Test classification report:")
print(classification_report(test_labels, test_preds, target_names=label_names, zero_division=0))
```
